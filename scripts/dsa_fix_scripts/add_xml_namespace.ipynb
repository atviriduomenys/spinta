```python
import pandas as pd

# Naudotojas turi nurodyti įvesties ir išvesties CSV failus
INPUT_CSV: str = ""  # pvz. "manifest-soap-small.csv"
OUTPUT_CSV: str = ""  # pvz. "manifest-soap-small-with-namespaces.csv"

if not INPUT_CSV or not OUTPUT_CSV:
    raise ValueError("Prašome nurodyti INPUT_CSV ir OUTPUT_CSV reikšmes.")

# Įkeliam CSV failą
df = pd.read_csv(INPUT_CSV)


# Funkcija pridedanti namespace prefiksą
def add_namespace_prefix(value: str | float | None, prefix: str = "jar:") -> str | float | None:
    if pd.isna(value):
        return value

    value = str(value)

    # Jeigu prasideda su '@', paliekam be pakeitimų
    if value.startswith("@"):
        return value

    # Pridedame prefiksą pradžioje, jeigu dar nėra
    if not value.startswith(prefix):
        value = prefix + value

    # Pridedame prefiksą po kiekvieno '/' simbolio
    parts = value.split("/")
    result = parts[0]

    for part in parts[1:]:
        if part and not part.startswith(prefix) and not part.startswith("@"):
            result += "/" + prefix + part
        else:
            result += "/" + part

    return result


# Pritaikom transformaciją stulpeliui 'source'
df["source"] = df["source"].apply(lambda x: add_namespace_prefix(x))

# Parodome pavyzdį prieš/po
original_df = pd.read_csv(INPUT_CSV)
comparison = pd.DataFrame({"Original": original_df["source"], "Modified": df["source"]})

# Parodome tik eilutes, kuriose yra pakeitimų
comparison = comparison[comparison["Original"] != comparison["Modified"]]
print("Palyginimas (prieš ir po pakeitimų):")
print(comparison.head(10))

# Išsaugome rezultatą į naują CSV failą
df.to_csv(OUTPUT_CSV, index=False)
print(f"\nPakeitimai išsaugoti į '{OUTPUT_CSV}'")
```
